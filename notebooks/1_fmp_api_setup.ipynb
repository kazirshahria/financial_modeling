```python
from fmp_api.fmp import FMP
import pandas as pd
```

```python
fmp = FMP()
```

```python
tickers = ['GPIX', 'GPIQ', 'DTCR', 'JEPI', 'JEPQ', 'SCHD']
etf_infos = []

for ticker in tickers:
    etf_info = fmp.etf_information(ticker)
    if etf_info:
        etf_infos.append(etf_info[0])

df = pd.DataFrame(etf_infos)
df
```
